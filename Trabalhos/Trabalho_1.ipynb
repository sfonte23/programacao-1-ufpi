3. Uma revendedora de carros usados paga a seus funcionários vendedores um salário fixo por mês, mais uma comissão também fixa para cada carro vendido e mais 5% do valor das vendas por ele efetuadas. Escrever um algoritmo que leia o número de carros por ele vendidos, o valor total de suas vendas, o salário fixo e o valor que ele recebe por carro vendido. Calcule e escreva o salário final do vendedor.

```python
ComissaoValorVendido= 0.05 #comissão do valor total do carro

CarrosVendidos = int(input("Quantos carros foram vendidos? "))
ValorTotaldeVenda= float(input("Qual foi o valor total de venda? "))
Salario = float(input("Qual o salario fixo? "))
ComissaoCarrosVendidos = float(input("Qual a comissão paga por carro vendido? "))

print(f'O salário final do vendedor será de R$ {Salario+(CarrosVendidos*ComissaoCarrosVendidos)+(ValorTotaldeVenda*ComissaoValorVendido)}')
```

4. Escreva um algoritmo para ler uma temperatura em graus Fahrenheit, calcular e escrever o valor correspondente em graus Celsius

```python
#Conversão de Fahrenheit para Celsius: (F-32)*5)/9

TemperaturaF = float(input("Escreva a temperatura em Fahrenheit: "))
TemperaturaC = ((TemperaturaF-32)*5)/9
print(f'A temperatura em Celsius é {TemperaturaC:,.2f}º C')
```

5. No mercadinho Mamão com Açucar as maçãs custam R$3,50 cada se forem compradas menos de uma dúzia, e R$3,00 se forem compradas pelo menos 12. Escreva um programa que calcule e escreva o custo total da compra.

```python
QuantidadeComprada = int(input("Digite a quantidade de Maçãs compradas: "))

if QuantidadeComprada < 12:
  print(f'O valor de sua compra foi de R${QuantidadeComprada*3.5}')

else:
  print(f'O valor de sua compra foi de R${QuantidadeComprada*3}')
```
